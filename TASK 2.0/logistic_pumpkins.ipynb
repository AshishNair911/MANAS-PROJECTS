```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train = pd.read_csv("/Users/maila/OneDrive/Desktop/LOGISTIC REGRESSION/Gotem_Pumpkins.csv")
train
```

```python
train = pd.get_dummies(train, columns = ['Class'],drop_first = True)
train
```

```python
numeric_columns_1 = train.select_dtypes(include=['int64', 'float64']).columns
#HOT ENCODING ALL INTEGER AND FLOAT VALUES TO VALUES BETWEEN 0 AND 1
for col in numeric_columns_1:
    min_val = train[col].min()
    max_val = train[col].max()
    train[col] = (train[col] - min_val) / (max_val - min_val)
train
```

```python
test = pd.read_csv("/Users/maila/OneDrive/Desktop/LOGISTIC REGRESSION/Freyja_Pumpkins.csv")
test
```

```python
numeric_columns_2 = test.select_dtypes(include=['int64', 'float64']).columns
#HOT ENCODING ALL INTEGER AND FLOAT VALUES TO VALUES BETWEEN 0 AND 1
test = pd.get_dummies(test, columns = ['Class'],drop_first = True)
for col in numeric_columns_2:
    min_val = test[col].min()
    max_val = test[col].max()
    test[col] = (test[col] - min_val) / (max_val - min_val)
test
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
# Cost function
def compute_cost(X, y, weights):
    m = len(y)
    h = sigmoid(np.dot(X, weights))
    cost = (-1 / m) * (np.dot(y.T, np.log(h)) + np.dot((1 - y).T, np.log(1 - h)))
    return cost
# Gradient descent Function
def gradient_descent(X, y, weights, learning_rate, iterations):
    m = len(y)
    for _ in range(iterations):
        h = sigmoid(np.dot(X, weights))
        gradient = np.dot(X.T, (h - y)) / m
        weights -= learning_rate * gradient
    return weights
#Applying the one vs all logistic regression model i.e considering a single class as one class and every other class combined as another class
def train_one(X, y, num_classes, learning_rate, iterations):
    m, n = X.shape
    all_weights = np.zeros((num_classes, n))
    for c in range(num_classes):
        # Create binary labels for class `c`
        y_binary = np.where(y == c, 1, 0).reshape(-1, 1)
        weights = np.zeros((n, 1))
        weights = gradient_descent(X, y_binary, weights, learning_rate, iterations)
        all_weights[c, :] = weights.flatten()
    return all_weights
# function for multiclass logistic regression

def predict_multiclass(X, all_weights):
    probabilities = sigmoid(np.dot(X, all_weights.T))
    return np.argmax(probabilities, axis=1)
```

```python
# Assuming the last column is the target variable in the training data
X_train = train.iloc[:, :-1].values  
y_train = train.iloc[:, -1].values
X_test = test.values
```

```python
#Add a bias term (column of 1s) to the feature matrices meaning
#Adding a column of ones ensures that the intercept is included as part of the matrix multiplication in the model.
train_columns = train.iloc[:, :-1].columns
X_test = test[train_columns].values  # Align test data with training features
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"all_weights shape: {all_weights.shape}")
```

```python
#Initializing parameters for the functions defined
num_classes = len(np.unique(y_train))  # Number of unique classes
learning_rate = 0.01
iterations = 1000
```

```python
#Using the one vs all logistic regression model
all_weights = train_one(X_train, y_train, num_classes, learning_rate, iterations)
```

```python
predictions = predict_multiclass(X_test, all_weights)
```

```python
y_test = test.iloc[:, -1].values
#Calculate accuracy
correct_predictions = np.sum(predictions == y_test)
total_predictions = len(y_test)
accuracy = correct_predictions / total_predictions
print("Accuracy is",accuracy*100, "%")
```
